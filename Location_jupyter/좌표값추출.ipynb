```python
count = 0

def rmduplicated(memlist):
    for i, row in enumerate(memlist):
        if memlist[i-1][0] == row[0]:
            avg = (memlist[i-2][2] + memlist[i-3][2] + memlist[i-4][2] + memlist[i-5][2] + memlist[i-6][2])/5

            if abs(memlist[i-1][2] - avg) < abs(row[2] - avg):
                memlist.remove(row) #i행 삭제
            else:
                memlist.remove(memlist[i-1]) #i-1행 삭제

rmduplicated(tzuyu)
rmduplicated(nayeon)
rmduplicated(jeongyeon)
rmduplicated(dahyun)
rmduplicated(sana)
rmduplicated(jihyo)
rmduplicated(momo)
rmduplicated(mina)
rmduplicated(chaeyoung)
```

```python
# 튀는값 제거 
def moving_avg_frame(df): 
    for idx, row in df.iterrows():
        left_idx = 0
        right_idx = 0
        
        x = [] # 에러 생길경우 맨 뒤에 remove추가하고 for문 위로 위치 변경하면 됨 
        y = []
        w = []
        h = [] # 왼쪽 정렬된 값 idx-1 하나와 오른쪽값하나
        
        L_x = []
        L_y = []
        L_w = []
        L_h = [] # 오른쪽 정렬 안된값 1~3개 2frame간격까지 허용, 위치 여기 or 아래 for문
        
        if idx < 1: # 1 frame부터
            pass
        
        elif idx == len(df)-2: # 1 frame부터
            break

        x.append(df.iloc[idx-1,2])
        y.append(df.iloc[idx-1,3])
        w.append(df.iloc[idx-1,4])
        h.append(df.iloc[idx-1,5]) # 루프당 초기값
        
        left_idx = idx - 1
            #
        if row["w"] == 0: # frame이 비었으면
            R_idx_list = [] # error 2 - 밑에서 위로 옮김
            for i in range(len(df)-(idx+1)): # 전체 길이에서 지금까지 진행된 index 값을 빼서 돌림. (뒤에인덱스만 돌린다.)
                idx_w = 0
                if df.iloc[idx+i+1, 4] != 0: # 만약 0이 아니면 리스트에 값을 넣는다.
                    idx_w = idx+i+1
                    L_x.append(df.iloc[idx_w,2]) # idx+1, idx+2, idx+3 값이 2 frame건너 뛰는거 까지 허용
                    L_y.append(df.iloc[idx_w,3])
                    L_w.append(df.iloc[idx_w,4])
                    L_h.append(df.iloc[idx_w,5])
                    R_idx_list.append(idx_w)

                    # error 1 - 들여쓰기 4 space 땡김
                    for j in range(1,6): # 값이 있는 frame부터 2frame interval까지 허용하여 최대 3개의 값을 list에 저장하는 코드
                        if idx_w + j == len(df) - 1: # 끝까지 돌앗을때 멈춤
                            break
                        
                        elif df.iloc[idx_w + j,4] != 0:
                            if len(L_x) == 3 or (len(L_x) < 2 and j == 3):
                                break

                            L_x.append(df.iloc[idx_w + j,2]) # idx+1, idx+2, idx+3 값이 2 frame건너 뛰는거 까지 허용
                            L_y.append(df.iloc[idx_w + j,3])
                            L_w.append(df.iloc[idx_w + j,4])
                            L_h.append(df.iloc[idx_w + j,5])
                            R_idx_list.append(idx_w + j) # 저장된 배열의 index값
                            
                        else:
                            pass 
                    break 
            
            print(L_x,L_y,L_w,L_h)
        
            # 값 list의 값중 하나를 추출
            if len(L_x) == 1: # 첫번째 값 사용
                x.append(L_x[0])
                y.append(L_y[0])
                w.append(L_w[0])
                h.append(L_h[0])
                right_idx = R_idx_list[0]
                
            elif len(L_x) == 2:
                if abs(x[0] - L_x[0]) < abs(x[0] - L_x[1]): # 왼쪽값과 차이값 작은것으로 적용
                    x.append(L_x[0])
                    y.append(L_y[0])
                    w.append(L_w[0])
                    h.append(L_h[0])
                    right_idx = R_idx_list[0]
                
                else:
                    x.append(L_x[1])
                    y.append(L_y[1])
                    w.append(L_w[1])
                    h.append(L_h[1])
                    right_idx = R_idx_list[1]
                
            elif len(L_x) == 3:
                if abs(L_x[0]-L_x[1]) < abs(L_x[1]-L_x[2]) and abs(L_x[0]-L_x[1]) < abs(L_x[2]-L_x[0]):
                    x.append(L_x[0])
                    y.append(L_y[0])
                    w.append(L_w[0])
                    h.append(L_h[0])  
                    right_idx = R_idx_list[0]
                
                elif abs(L_x[1]-L_x[2]) < abs(L_x[0]-L_x[1]) and abs(L_x[1]-L_x[2]) < abs(L_x[2]-L_x[0]):
                    x.append(L_x[1])
                    y.append(L_y[1])
                    w.append(L_w[1])
                    h.append(L_h[1])  
                    right_idx = R_idx_list[1]
                    
                else:
                    x.append(L_x[2])
                    y.append(L_y[2])
                    w.append(L_w[2])
                    h.append(L_h[2]) 
                    right_idx = R_idx_list[2]

            zero_len = right_idx - left_idx # 0의 길이
            
            if len(x) == 1:
                return
            
            div_val_x = (x[1] - x[0])/zero_len # 차이를 왼쪽 오른쪽 인덱스 길이만큼 나눈값
            div_val_y = (y[1] - y[0])/zero_len
            div_val_w = (w[1] - w[0])/zero_len
            div_val_h = (h[1] - h[0])/zero_len
            
            #선형 증가값 적용
            for k in range(zero_len):
                df.iloc[idx+k, 2] = x[0] + div_val_x*(k+1)
                df.iloc[idx+k, 3] = y[0] + div_val_y*(k+1)
                df.iloc[idx+k, 4] = w[0] + div_val_w*(k+1)
                df.iloc[idx+k, 5] = h[0] + div_val_h*(k+1)
                
            for s in range(zero_len): # 선형증가한 위치만큼 뒤로 이동
                pass
                

moving_avg_frame(cf1)
moving_avg_frame(cf2)
moving_avg_frame(cf3)
moving_avg_frame(cf4)
moving_avg_frame(cf5)
moving_avg_frame(cf6)
moving_avg_frame(cf7)
moving_avg_frame(cf8)
moving_avg_frame(cf9)
```
